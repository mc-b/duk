# Jobs

Jobs sind einmalig ausgeführte Aufgaben, die in einer isolierten Umgebung laufen und nach Abschluss beendet werden. Sie eignen sich für klar abgegrenzte, endliche Arbeitsprozesse wie Datenverarbeitung, Batch-Berechnungen, das Versenden von E-Mails oder das Rendern von Inhalten.

Ein Job kann mehrere parallele Ausführungen starten, um unabhängige, aber zusammengehörige Aufgaben effizient abzuarbeiten. Dabei liegt der Fokus auf der vollständigen Abarbeitung einer definierten Aufgabe, nicht auf Wiederholung oder Zeitsteuerung.

Jobs sind nicht für dauerhaft laufende Prozesse oder kontinuierliche Hintergrunddienste gedacht. Ebenso sind sie ungeeignet für stark gekoppelte, intensiv kommunizierende parallele Systeme. 

Durch die Angabe von `parallelism` (Gleichzeitigkeit), und `completions` (Anzahl Jobs) können Jobs aufgeteilt und parallel gestartet werden. Ohne dabei den Cluster unnötig zu belasten.

---

Das nachfolgende Beispiel aus [Jobs Konzept](https://kubernetes.io/docs/concepts/workloads/controllers/job/), startet jeweils 2 parallele Pods und stellt sicher, dass 6 Pods durchlaufen.

```python
%%bash
kubectl apply -f - <<EOF
apiVersion: batch/v1
kind: Job
metadata:
  name: pi
spec:
  parallelism: 2
  completions: 6
  template:
    spec:
      containers:
      - name: pi
        image: perl:5.34.0
        command: ["perl",  "-Mbignum=bpi", "-wle", "print bpi(2000)"]
      restartPolicy: Never
  backoffLimit: 4
EOF
```

Anzeigen der Jobs/Pods und deren Ausgabe

```python
%%bash
kubectl get jobs
kubectl get jobs/pi --watch
```

---
### Aufräumen

```python
%%bash
kubectl delete job pi
```
