# Cron Jobs

Cron Jobs sind zeitgesteuerte Aufgaben, die automatisch in festgelegten Intervallen ausgeführt werden. Sie eignen sich besonders für wiederkehrende, batchorientierte Prozesse wie Backups, Log-Rotation, das Versenden von Reports oder periodische Datenverarbeitung.

Im Gegensatz zu einmalig gestarteten Jobs laufen Cron Jobs nach einem definierten Zeitplan (z. B. jede Stunde, täglich oder wöchentlich). Sie führen dabei unabhängig voneinander Aufgaben aus, ohne dass eine kontinuierliche Interaktion zwischen den einzelnen Ausführungen erforderlich ist.

Cron Jobs sind jedoch nicht für langlaufende, dauerhaft aktive Prozesse oder für stark miteinander kommunizierende parallele Systeme gedacht. 

---
Beispiel für einen CronJobs aus [Run Jobs](https://kubernetes.io/docs/tasks/job/):

```python
%%bash
kubectl apply -f - <<EOF
apiVersion: batch/v1
kind: CronJob
metadata:
  name: hello
spec:
  schedule: "* * * * *"
  jobTemplate:
    spec:
      template:
        spec:
          containers:
          - name: hello
            image: busybox:1.28
            imagePullPolicy: IfNotPresent
            command:
            - /bin/sh
            - -c
            - date; echo Hello from the Kubernetes cluster
          restartPolicy: OnFailure
EOF
```

Anzeigen der Jobs/Pods und deren Ausgabe

```python
%%bash
kubectl get cronjobs
kubectl get jobs --watch
```

---
### Aufräumen

```python
%%bash
kubectl delete cronjob hello
```
