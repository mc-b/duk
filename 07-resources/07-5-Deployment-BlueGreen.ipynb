Übung: Blue/Green Deployments
-----------------

In dieser Übung führen wir einen Blue/Green Deployment durch.

Dazu werden zwei Deployments von Webshop erstellt und mittels Veränderung der Zusteuerung des Service zwischen den Version hin und her geschaltet.

Das passiert in einer eigenen Namespace um die Resultate gezielt Darstellen zu können:

```python
! kubectl create namespace bluegreen
```

Deployment und Services wurde in einzelne Dateien aufgeteilt. Die Ingress Ressource, aus Einfachheitsgründen weggelassen.

Die Labels und das Image wurden um eine Version Angabe erweitert:

    apiVersion: v1
    kind: Service
    metadata:
      name: webshop
      labels:
        app: webshop
    spec:
      ...
      selector:
        app: webshop
        version: "1.0.0"    
        
- - -

    apiVersion: apps/v1
    kind: Deployment
    metadata:
      name: webshop
      version: 1.0.0
    ...
    spec:
      containers:
      - name: webshop
        image: registry.gitlab.com/mc-b/misegr/misegr/webshop:1.0.0
        imagePullPolicy: IfNotPresent       



```python
! kubectl apply -f 07-5-Deployment/webshop-deployment-blue.yaml --namespace bluegreen
! kubectl apply -f 07-5-Deployment/webshop-service-blue.yaml --namespace bluegreen
```

Das Ergebnis unterscheidet sich, vorerst, nicht von einem normalen Deployment

```python
! kubectl get all --namespace bluegreen
! echo "http://$(cat ~/data/server-ip):$(kubectl get service --namespace bluegreen webshop -o=jsonpath='{ .spec.ports[0].nodePort }')"
```

- - -

Für das **green** Deployment erstellen wir nur ein Deployment aber kein Service:

```python
! kubectl apply -f 07-5-Deployment/webshop-deployment-green.yaml --namespace bluegreen
! kubectl get all --namespace bluegreen -o wide
```

### Vorbereitung für das Blue/Green Deployment

Bevor wir den **Selector des Services** ändern, richten wir **Headlamp** wie folgt ein:

1. Rechts die Ansicht **Map** auswählen.
2. Den Namespace `bluegreen` aktivieren und alle anderen Namespaces deaktivieren.
3. **Expand All** auswählen.

So können wir das Ergebnis anschliessend visuell überprüfen.


```python
%%bash
echo "HeadLamp: http://"$(cat ~/data/server-ip)":30444"
kubectl create token headlamp-admin -n kube-system --duration=48h
```

Der Service leitet den Traffic aktuell an die **blue** Pods weiter.

Erst nachdem der Selector des Services auf die **green** Version geändert wurde, ist diese für die Anwender sichtbar.

        apiVersion: v1
        kind: Service
        metadata:
          name: webshop
          labels:
            app: webshop
        spec:
          ...
          selector:
            app: webshop
            version: "2.0.0"
        

```python
! kubectl apply -f 07-5-Deployment/webshop-service-green.yaml --namespace bluegreen
```

An den Anzahl Ressourcen ändert der Selector nichts

```python
! kubectl get all --namespace bluegreen -o wide
```

```python
! kubectl delete -f 07-5-Deployment/webshop-deployment-blue.yaml --namespace bluegreen
! kubectl get all --namespace bluegreen
```
