# Open Telemetry

Die OpenTelemetry-Demo besteht aus Microservices, die in verschiedenen Programmiersprachen geschrieben sind und über gRPC und HTTP miteinander kommunizieren; sowie einem Lastgenerator, der [Locust](https://locust.io/) verwendet , um Benutzerverkehr zu simulieren


```python
%%bash
helm repo add open-telemetry https://open-telemetry.github.io/opentelemetry-helm-charts
helm install my-otel-demo open-telemetry/opentelemetry-demo --create-namespace -n otel-demo
```

```python
%%bash
echo "HeadLamp: http://"$(cat ~/data/server-ip)":30444"
kubectl create token headlamp-admin -n kube-system --duration=48h
kubectl -n otel-demo patch service frontend-proxy -p '{"spec": {"type": "LoadBalancer"}}'
echo "Webstore          : http://"$(cat ~/data/server-ip)":"$(kubectl get service --namespace otel-demo frontend-proxy -o=jsonpath='{ .spec.ports[0].nodePort }')
echo "Jaeger UI         : http://"$(cat ~/data/server-ip)":"$(kubectl get service --namespace otel-demo frontend-proxy -o=jsonpath='{ .spec.ports[0].nodePort }')/jaeger/ui
echo "Grafana           : http://"$(cat ~/data/server-ip)":"$(kubectl get service --namespace otel-demo frontend-proxy -o=jsonpath='{ .spec.ports[0].nodePort }')/grafana
echo "Load Generator UI : http://"$(cat ~/data/server-ip)":"$(kubectl get service --namespace otel-demo frontend-proxy -o=jsonpath='{ .spec.ports[0].nodePort }')/loadgen
echo "Feature Flags UI  : http://"$(cat ~/data/server-ip)":"$(kubectl get service --namespace otel-demo frontend-proxy -o=jsonpath='{ .spec.ports[0].nodePort }')/feature
```
